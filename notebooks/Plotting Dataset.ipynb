```python
import pandas as pd
import seaborn as sns
```

```python
frame = pd.read_csv("dataset.csv")
```

```python
frame.describe()
```

```python
sns.set(style="darkgrid")
```

```python
sns.lineplot(x="step", y="rmp",
             hue="cell",
             data=frame)
```

```python
sns.lineplot(x="step", y="rheobase",
             hue="cell",
             data=frame)
```

```python
ax= sns.lineplot(x="step", y="firstOver20Value",
            hue="cell",
             data=frame)
ax.set(ylim=(-50, -60))
ax.set_ylabel("AP Threshold [mv]")
ax
```

```python
ax= sns.lineplot(x="step", y="res_maxmean",
            hue="cell",
             data=frame)
ax.set_ylabel("Series Resistance [MOhm]")
```

```python
ax= sns.lineplot(x="step", y="res_stablemean",
            hue="cell",
             data=frame)
```

```python
ax= sns.lineplot(x="step", y="ih-slope",
            hue="cell",
             data=frame)
ax.set_ylabel("Ih Slope")
ax
```

```python
ax= sns.lineplot(x="step", y="maxSlopeValue",
            hue="cell",
             data=frame)

ax.set(ylim=(300, 500))
ax.set_ylabel("AP max rise dv/dt")
ax
```

```python
ax= sns.lineplot(x="step", y="peakValue",
            hue="cell",
             data=frame)

ax.set(ylim=(30, 35))
ax.set_ylabel("AP peak Value [mv]")
ax
```
